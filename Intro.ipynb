```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def import_dta(link):
    name = pd.read_stata(link)
    return name
```

```python
dax = import_dta("C:\\Users\\gonza\\Documents\\Python\\TSA\\dax.dta")
```

```python
dax.head()
```

```python
dax.dtypes
```

```python
dax.describe(include="all")
```

```python
fig, ax = plt.subplots(figsize = (18,12))
ax.plot(dax['time'], dax['dax'])
```

```python
dax = dax.set_index("time")
```

```python
fy = dax.resample("M").sum()
fig, ax = plt.subplots(figsize = (18,12))
ax.plot(fy['dax'])
```

```python
fy = dax.resample("Q").sum()
fig, ax = plt.subplots(figsize = (18,12))
ax.plot(fy['dax'])
```
